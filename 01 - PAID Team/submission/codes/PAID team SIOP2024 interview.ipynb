# Zero-shot prompting using GPT-4

Instead of directly using the original test set data format, we transformed the original dataset into a new format. In the transformed dataset, each question and response was extracted from the "question_response" column. Now we have 'Q1', 'Q2' etc. for questions and 'R1', 'R2' etc. for responses. This new dataset was saved as "interview_test_converted.csv". Which is the data we used in the following steps.

```python
# Save the data file path
# You may need to change the path. You can find the dataset under 'data' folder.
transformed_test_file_path = '/content/drive/MyDrive/01_Research/2024 SIOP competition/Reproduce/data/interview_test_converted.csv'
```

```python
import pandas as pd

# Import data files
ds_test = pd.read_csv(transformed_test_file_path)
```

```python
ds_test
```

```python
# Chat completion using for loop
predict_test = []
for i in range(len(ds_test)):
  completion = client.chat.completions.create(
      model="gpt-4-0125-preview",
      messages= [{'role': 'system', 'content': """Act as a job candidate in a job interview"""},
             {'role': 'user', 'content': f"""Respond to the last question using no more than 120 words. Keep the tone and personality reflected in the provided previous responses.
Question: {ds_test['Q1'][i]}
Response: {ds_test['R1'][i]}

Question: {ds_test['Q2'][i]}
Response: {ds_test['R2'][i]}

Question: {ds_test['Q3'][i]}
Response: {ds_test['R3'][i]}

Question: {ds_test['Q4'][i]}"""}],
      temperature = 1,
      seed = 666666
    )
  predict_test.append(completion.choices[0].message.content)
  print(completion.choices[0].message.content)
```
